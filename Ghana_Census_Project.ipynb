# Research Question

**"How does access to post-secondary education make a positive impact career prospects in Ghana?"**



# Problem Statement

With the increasing development of services and technology globally, higher education attainment has shown in many studies to correlate to employability and higher salaries. As the services growth rate continues to increase in Ghana, it is pertinent for Ghanaian youth to have access to quality education and skills training to increase their socio-economic status as country focuses more on white collar professional roles and less on agrarian labor. Our research aims to understand how does access to post-secondary education make a positive impact career prospects in Ghana?

```python
print(df_gen_ed.head(15))
print(df_employment.head(15))
```

```python
#Clean up columns
for df in [df_ed_career, df_employment]:
    df['hhmid'] = df['hhmid'].astype(int)
    df['age'] = df['age'].astype(int)
    df['interviewedid'] = df['interviewedid'].str.strip()

#Merge on hhmid, age, and interviewedid
df_merged = pd.merge(df_ed_career, df_employment,
                     on=['hhmid', 'age', 'interviewedid'],
                     how='inner')  # Use 'inner' to only keep people in both datasets

#Check what columns are present in the merged dataframe
print(df_merged.columns.tolist())

#Check the merged dataframe
df_merged[['age', 'highestachieved', 'maintrade']].head()

# Filter out null values for 'highestachieved' and 'maintrade'
df_filtered = df_merged[['highestachieved', 'maintrade']].dropna()

# Count how many people per job per education level
education_trade_counts = df_filtered.groupby(['highestachieved', 'maintrade']).size().reset_index(name='count')
ed_trade_count_filt = education_trade_counts[education_trade_counts['count']>30]

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=ed_trade_count_filt, x='highestachieved', y='count', hue='maintrade')
plt.title("Distribution of Main Trades by Highest Grade Completed")
plt.xlabel("Highest Grade")
plt.ylabel("Number of Individuals")
plt.legend(title='Main Trade', loc='upper left')
plt.show()
```

```python
print(df_merged.columns.tolist())
```

```python
# Replace 'monthlyincome' with your actual column name
df_income = df_merged[['highestachieved', 'paidamount']].dropna()

# Ensure salary is numeric
df_income['paidamount'] = pd.to_numeric(df_income['paidamount'], errors='coerce')
df_income = df_income.dropna(subset=['paidamount'])  # Drop rows where conversion failed
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_income, x='highestachieved', y='paidamount', palette='Set3')
plt.title("Monthly Income by Highest Grade Completed")
plt.xlabel("Highest Grade Completed")
plt.ylabel("Monthly Income (GHS)")
plt.xticks(rotation=45)
plt.ylim(0, 5000)
plt.grid(True)
plt.show()
```

Monthly income has been shown to be positively correlated to educational attainment beyond the Higher National Diploma of Ghana, which is the equivalent of an American Associates degree.
