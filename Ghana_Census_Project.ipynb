# Research Question

**"How does access to post-secondary education make a positive impact career prospects in Ghana?"**



# Problem Statement

With the increasing development of services and technology globally, higher education attainment has shown in many studies to correlate to employability and higher salaries. As the services growth rate continues to increase in Ghana, it is pertinent for Ghanaian youth to have access to quality education and skills training to increase their socio-economic status as country focuses more on white collar professional roles and less on agrarian labor. Our research aims to understand how does access to post-secondary education make a positive impact career prospects in Ghana?
